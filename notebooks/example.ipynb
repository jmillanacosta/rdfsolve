## RDF-Config approach



Test: NanoSafetyRDF

```python
ns = RDFConfigr(
    endpoint="https://nanosafety.rdf.bigcat-bioinformatics.org/sparql/",
    path="/rdfconfig",
    void_iri="https://nanosafety.rdf.bigcat-bioinformatics.org/sparql/",
    dataset_name="nanosafety_rdf"
)
print(
    "graph IRIs:", ns._graph_iris
)  # You can now decide which graph iris to keep and set this value with wp._graph_iris = list()
ns.get_config()
```

```python
print(ns.config)
```
